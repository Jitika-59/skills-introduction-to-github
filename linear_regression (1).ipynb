```python
def loss_function(m, b, points):
    total_error = 0
    for i in range(len(points)):
        x = points.iloc[i].YearsExperience
        y = points.iloc[i].Salary
        total_error += (y - (m * x + b)) ** 2
    total_error / float(len(points))   
#function is to calculate loss mannually
```

```python
def gradient_descent(m_now, b_now,points,L):
    m_gradient = 0
    b_gradient = 0
    n = len(points)
    for i in range(n):
        x = points.iloc[i].YearsExperience
        y = points.iloc[i].Salary
        # Gradient formulas
        m_gradient += -(2/n) * x * (y - (m_now * x + b_now))
        b_gradient += -(2/n) * (y - (m_now * x + b_now))
     # Update step
    m = m_now - m_gradient * L
    b = b_now - b_gradient * L
    return m, b
# Initialize parameters
m = 0
b = 0
L = 0.0001
epochs = 1000
# Training loop
for i in range(epochs):
    if i % 50 == 0:
        print(f"Epoch:{i}")
    m, b = gradient_descent(m, b, data, L)
# Final result
print(f"\nFinal Model: Salary = {m:.2f} * YearsExperience + {b:.2f}")

#use epochs in linear regression 
#(with gradient descent) to iteratively train the model by updating its 
#parameters multiple times across the dataset, helping it gradually improve 
#and reach optimal performance.
plt.scatter(data.YearsExperience, data.Salary, color="black")
```
